# Histogram ve Yoğunluk Grafikleri

```python
# Histogram ve yoğunluk grafikleri sayısal değişkenler için kullanılır
import seaborn as sns
diamonds = sns.load_dataset("diamonds")
df = diamonds.copy()
df.head()
```

```python
#kde yoğunluk temsili
sns.distplot(df["price"], kde=False);
```

```python
#bins sütun sayısı diyebiliriz
sns.distplot(df["price"], bins=10, kde=False);
```

```python
sns.distplot(df["price"]);
```

```python
#Sadece yoğunluk görmek için
sns.distplot(df["price"], hist=False);
```

```python
#Grafiğin altını boyama
sns.kdeplot(df.price, fill=True);
```

## Çaprazlamalar

```python
import seaborn as sns
diamonds = sns.load_dataset("diamonds")
df = diamonds.copy()
df.head()
```

```python
sns.kdeplot(df.price, fill=True);
```

```python
(sns.FacetGrid(df,
              hue = "cut",
              height = 5,
              xlim = (0, 10000))
.map(sns.kdeplot, "price", fill=True)
.add_legend()
);
```

```python
sns.catplot(x="cut", y="price", hue="color", kind="point", data=df);
```
